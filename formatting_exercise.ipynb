```python
#Import required packages
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests
import numpy as np

np.random.seed(613)
x1=np.arange(50)
y1=np.random.randint(0,75,50)
x2=np.array(["Luff","Zoro","Nami","Usopp","Sanji"])
y2=np.array([110,180,240,99,220])
```

```python
# Plot the 2 graphs
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize = (7,3))

ax1.scatter(x1,y1)
ax2.bar(x2,y2)
```

```python
# Experiment with formatting
fig, (ax1, ax2) = plt.subplots(ncols=2,
                               nrows=1,
                               figsize = (7,3),
                               )

ax1.scatter(x1,
            y1,
            color = "orange")
plt.sca(ax1) # Apply formatting to scatterplot
plt.xticks(rotation=45, ha='right')
plt.xlabel('Test scores', fontsize=18, fontfamily = 'cursive', color = 'orange')

ax2.bar(x2,
        y2,
        color = "orange")
plt.sca(ax2) # Apply formatting to bar graph
plt.xticks(rotation=45, ha='right')
plt.xlabel('Names', fontsize=18, fontfamily = 'cursive', color = 'orange')





```
