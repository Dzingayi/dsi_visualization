```python
#This visualization is appealing (for my specific use case) because it shows the size of each transaction block

#Import libraries
!pip install pywaffle
import matplotlib.pyplot as plt

from pywaffle import Waffle

# create simple dummy data
data = {'Block 350,000': 3, 'Block 350,001': 2, 'Block 350,002': 1}

#Percentages of each block
percentage = [f"{k} ({int(v / sum(data.values()) * 100)}%)" for k, v in data.items()]

# Basic waffle
plt.figure(
  FigureClass=Waffle,
  rows=5,
  columns=20,
  values=data,
  labels = percentage,
  legend={'loc': 'upper left', 'bbox_to_anchor': (1.05, 1)},
)
#Add title
plt.title("Latest BTC Blocks")

#Display chart
plt.show()
```
